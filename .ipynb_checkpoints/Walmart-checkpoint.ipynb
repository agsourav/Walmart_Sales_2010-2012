## Weekly sales across each dept across each store

### segregating by stores

```python
df_stores = df.groupby('Store')
```

```python
mean_sales_stores = df_stores['Weekly_Sales'].mean()
```

```python
mean_sales_stores
```

```python
plt.figure(figsize = (12,10))
sns.barplot(x = mean_sales_stores.index, y = mean_sales_stores.values)
```

```python
mean_sales_overall = np.mean(mean_sales_stores)
median = np.median(mean_sales_stores)
print('overall mean sales: {0}\tmedian sales: {1}'.format(mean_sales_overall, median))
#selecting stores with mean sales less than global mean
```

```python
#mean sales of stores greater than global mean
df_stores_1 = mean_sales_stores[mean_sales_stores > mean_sales_overall]
df_stores_1.shape
```

```python
sns.barplot(x = df_stores_1.index, y = df_stores_1.values)
```

```python
store_19.groupby('Dept').mean().style.highlight_max(color = 'red', subset = ['Weekly_Sales']).highlight_min(color = 'green', subset = ['Weekly_Sales'])
```

```python
def color_red_negative(x):
    color = 'red' if x < 0 else 'white'
    return 'color:'+color
```
